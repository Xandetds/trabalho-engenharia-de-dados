# Populate Landing (setup local)

Faz upload dos CSVs brutos para o bucket `landing` no MinIO.  
**Use apenas para testes locais** — em produção quem faz isso é o Airflow (issue #5).

> `minio` já está disponível via startup script.

```python
import os

CSV_DIR = "/app/data/landing/tse"
BUCKET  = "landing"
PREFIX  = "tse"

arquivos = sorted(f for f in os.listdir(CSV_DIR) if f.endswith(".csv"))
print(f"{len(arquivos)} arquivos encontrados em {CSV_DIR}:\n")
for f in arquivos:
    print(f"  {f}")
```

```python
for arquivo in arquivos:
    caminho = os.path.join(CSV_DIR, arquivo)
    chave   = f"{PREFIX}/{arquivo}"
    tamanho = os.path.getsize(caminho) / 1024

    minio.upload_file(caminho, BUCKET, chave)
    print(f"  ✓ {chave} ({tamanho:.1f} KB)")

print("\nUpload concluído.")
```

```python
# Verificação: lista o que está no bucket
resp = minio.list_objects_v2(Bucket=BUCKET, Prefix=PREFIX)
print(f"=== Objetos em s3://{BUCKET}/{PREFIX} ===")
for obj in resp.get("Contents", []):
    print(f"  {obj['Key']}  ({obj['Size']/1024:.1f} KB)")
```
