# Bronze 

## Carregar os dados da Landing Zone

```python
caminho_landing = "s3a://landing/tse/"

df_bens              = spark.read.option("multiline", "true").json(caminho_landing + "bens.json")
df_bens_estatistica  = spark.read.option("multiline", "true").json(caminho_landing + "bens_estatistica.json")
df_candidatura       = spark.read.option("multiline", "true").json(caminho_landing + "candidatura.json")
df_cargo             = spark.read.option("multiline", "true").json(caminho_landing + "cargo.json")
df_instrucao         = spark.read.option("multiline", "true").json(caminho_landing + "instrucao.json")
df_municipio         = spark.read.option("multiline", "true").json(caminho_landing + "municipio.json")
df_ocupacao          = spark.read.option("multiline", "true").json(caminho_landing + "ocupacao.json")
df_partido           = spark.read.option("multiline", "true").json(caminho_landing + "partido.json")
df_situacao          = spark.read.option("multiline", "true").json(caminho_landing + "situacao.json")
df_tipo_bem          = spark.read.option("multiline", "true").json(caminho_landing + "tipo_bem.json")
```

## Adicionar colunas de Metadados (Data/Hora e Origem)

```python
from pyspark.sql.functions import current_timestamp, lit

df_bens              = df_bens.withColumn("data_hora_bronze", current_timestamp()).withColumn("fonte_dados", lit("landing.tse.bens"))
df_bens_estatistica  = df_bens_estatistica.withColumn("data_hora_bronze", current_timestamp()).withColumn("fonte_dados", lit("landing.tse.bens_estatistica"))
df_candidatura       = df_candidatura.withColumn("data_hora_bronze", current_timestamp()).withColumn("fonte_dados", lit("landing.tse.candidatura"))
df_cargo             = df_cargo.withColumn("data_hora_bronze", current_timestamp()).withColumn("fonte_dados", lit("landing.tse.cargo"))
df_instrucao         = df_instrucao.withColumn("data_hora_bronze", current_timestamp()).withColumn("fonte_dados", lit("landing.tse.instrucao"))
df_municipio         = df_municipio.withColumn("data_hora_bronze", current_timestamp()).withColumn("fonte_dados", lit("landing.tse.municipio"))
df_ocupacao          = df_ocupacao.withColumn("data_hora_bronze", current_timestamp()).withColumn("fonte_dados", lit("landing.tse.ocupacao"))
df_partido           = df_partido.withColumn("data_hora_bronze", current_timestamp()).withColumn("fonte_dados", lit("landing.tse.partido"))
df_situacao          = df_situacao.withColumn("data_hora_bronze", current_timestamp()).withColumn("fonte_dados", lit("landing.tse.situacao"))
df_tipo_bem          = df_tipo_bem.withColumn("data_hora_bronze", current_timestamp()).withColumn("fonte_dados", lit("landing.tse.tipo_bem"))
```

## Salvando no Bucket Bronze

```python
caminho_bronze = "s3a://bronze/tse/"

df_bens.write.format("delta").mode("overwrite").save(caminho_bronze + "bens")
df_bens_estatistica.write.format("delta").mode("overwrite").save(caminho_bronze + "bens_estatistica")
df_candidatura.write.format("delta").mode("overwrite").save(caminho_bronze + "candidatura")
df_cargo.write.format("delta").mode("overwrite").save(caminho_bronze + "cargo")
df_instrucao.write.format("delta").mode("overwrite").save(caminho_bronze + "instrucao")
df_municipio.write.format("delta").mode("overwrite").save(caminho_bronze + "municipio")
df_ocupacao.write.format("delta").mode("overwrite").save(caminho_bronze + "ocupacao")
df_partido.write.format("delta").mode("overwrite").save(caminho_bronze + "partido")
df_situacao.write.format("delta").mode("overwrite").save(caminho_bronze + "situacao")
df_tipo_bem.write.format("delta").mode("overwrite").save(caminho_bronze + "tipo_bem")

print("Todas as 10 tabelas foram processadas e salvas na camada Bronze com sucesso!")
```
